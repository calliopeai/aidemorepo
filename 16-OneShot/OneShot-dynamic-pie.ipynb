# Dynamic Pie Chart with Interactive Controls

This notebook creates an interactive pie chart with dynamic controls using plotly and ipywidgets. You can adjust the data, colors, and display options in real-time.

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display
```

## Sample Data

First, let's create some sample data that we can manipulate with our controls.

```python
# Create sample data
np.random.seed(42)
categories = ['Technology', 'Healthcare', 'Finance', 'Education', 'Retail', 'Manufacturing']
values = np.random.randint(10, 100, len(categories))

# Create DataFrame
df = pd.DataFrame({
    'Category': categories,
    'Value': values
})

print("Sample Data:")
print(df)
```

## Interactive Pie Chart Function

This function creates a pie chart that updates based on user controls.

```python
def create_dynamic_pie_chart(data_multiplier=1.0, hole_size=0.0, show_percentages=True, 
                             color_scheme='Set3', title_text='Dynamic Pie Chart',
                             min_value_filter=0):
    """
    Create a dynamic pie chart with customizable parameters
    """
    # Apply data transformations
    df_filtered = df.copy()
    df_filtered['Value'] = df_filtered['Value'] * data_multiplier
    
    # Filter by minimum value
    df_filtered = df_filtered[df_filtered['Value'] >= min_value_filter]
    
    # Create the pie chart
    fig = px.pie(df_filtered, 
                 values='Value', 
                 names='Category',
                 title=title_text,
                 color_discrete_sequence=px.colors.qualitative.__dict__[color_scheme],
                 hole=hole_size)
    
    # Update traces for percentage display
    if show_percentages:
        fig.update_traces(textposition='inside', textinfo='percent+label')
    else:
        fig.update_traces(textposition='inside', textinfo='label')
    
    # Update layout
    fig.update_layout(
        title_font_size=16,
        font_size=12,
        showlegend=True,
        width=600,
        height=500
    )
    
    fig.show()
```

## Interactive Controls

Use the sliders and options below to dynamically update the pie chart.

```python
# Create interactive widgets
data_multiplier_slider = widgets.FloatSlider(
    value=1.0,
    min=0.1,
    max=3.0,
    step=0.1,
    description='Data Multiplier:',
    style={'description_width': 'initial'}
)

hole_size_slider = widgets.FloatSlider(
    value=0.0,
    min=0.0,
    max=0.6,
    step=0.1,
    description='Hole Size:',
    style={'description_width': 'initial'}
)

percentage_checkbox = widgets.Checkbox(
    value=True,
    description='Show Percentages',
    style={'description_width': 'initial'}
)

color_scheme_dropdown = widgets.Dropdown(
    options=['Set3', 'Pastel', 'Dark2', 'Set1', 'Set2', 'Pastel1', 'Pastel2'],
    value='Set3',
    description='Color Scheme:',
    style={'description_width': 'initial'}
)

title_text_input = widgets.Text(
    value='Dynamic Pie Chart',
    description='Chart Title:',
    style={'description_width': 'initial'}
)

min_value_slider = widgets.IntSlider(
    value=0,
    min=0,
    max=50,
    step=5,
    description='Min Value Filter:',
    style={'description_width': 'initial'}
)
```

## Interactive Display

Run the cell below to display the interactive pie chart with all controls.

```python
# Create the interactive display
interactive_plot = interactive(
    create_dynamic_pie_chart,
    data_multiplier=data_multiplier_slider,
    hole_size=hole_size_slider,
    show_percentages=percentage_checkbox,
    color_scheme=color_scheme_dropdown,
    title_text=title_text_input,
    min_value_filter=min_value_slider
)

# Display the interactive plot
display(interactive_plot)
```

## Advanced Features

Let's add some additional features like data upload and real-time data generation.

```python
# Function to generate random data
def generate_random_data():
    global df
    categories = ['Technology', 'Healthcare', 'Finance', 'Education', 'Retail', 'Manufacturing']
    values = np.random.randint(10, 100, len(categories))
    df = pd.DataFrame({
        'Category': categories,
        'Value': values
    })
    print("New random data generated!")
    print(df)

# Create button for random data generation
random_data_button = widgets.Button(
    description='Generate Random Data',
    button_style='info',
    tooltip='Click to generate new random data'
)

def on_random_data_clicked(b):
    generate_random_data()

random_data_button.on_click(on_random_data_clicked)

display(random_data_button)
```

## Custom Data Input

You can also input your own data using the text areas below.

```python
# Custom data input widgets
categories_input = widgets.Textarea(
    value='Technology,Healthcare,Finance,Education,Retail,Manufacturing',
    placeholder='Enter categories separated by commas',
    description='Categories:',
    style={'description_width': 'initial'},
    layout={'width': '400px', 'height': '60px'}
)

values_input = widgets.Textarea(
    value='45,67,23,89,34,56',
    placeholder='Enter values separated by commas',
    description='Values:',
    style={'description_width': 'initial'},
    layout={'width': '400px', 'height': '60px'}
)

update_data_button = widgets.Button(
    description='Update Data',
    button_style='success',
    tooltip='Click to update the chart with custom data'
)

def update_custom_data(b):
    global df
    try:
        categories = [cat.strip() for cat in categories_input.value.split(',')]
        values = [float(val.strip()) for val in values_input.value.split(',')]
        
        if len(categories) != len(values):
            print("Error: Number of categories must match number of values")
            return
            
        df = pd.DataFrame({
            'Category': categories,
            'Value': values
        })
        print("Data updated successfully!")
        print(df)
    except Exception as e:
        print(f"Error updating data: {e}")

update_data_button.on_click(update_custom_data)

# Display custom data input widgets
display(widgets.VBox([
    categories_input,
    values_input,
    update_data_button
]))
```
