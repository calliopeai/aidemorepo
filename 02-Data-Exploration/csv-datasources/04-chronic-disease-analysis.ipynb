```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

plt.style.use('seaborn-v0_8-whitegrid')
DATA_PATH = Path.home() / 'lab' / 'data' / 'CSVs' / 'U.S._Chronic_Disease_Indicators.csv'
```

```python
df = pd.read_csv(DATA_PATH, low_memory=False)
print(f"Loaded {len(df):,} records")
df.head()
```

```python
# Topics covered
print("Health Topics:")
for topic in df['Topic'].unique():
    print(f"  - {topic}")
```

```python
# Diabetes prevalence by state (example indicator)
diabetes = df[(df['Topic'] == 'Diabetes') & 
              (df['Question'].str.contains('Prevalence', na=False)) &
              (df['DataValueType'] == 'Crude Prevalence')]

if len(diabetes) > 0:
    state_diabetes = diabetes.groupby('LocationDesc')['DataValue'].mean().sort_values(ascending=False).head(20)
    
    plt.figure(figsize=(12, 8))
    state_diabetes.plot(kind='barh', color='indianred')
    plt.title('Diabetes Prevalence by State (Top 20)')
    plt.xlabel('Prevalence Rate (%)')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
```
