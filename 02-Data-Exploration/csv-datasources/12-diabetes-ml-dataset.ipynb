```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

plt.style.use('seaborn-v0_8-whitegrid')
DATA_PATH = Path.home() / 'lab' / 'data' / 'CSVs' / 'diabetes_all_2016.csv'
```

```python
df = pd.read_csv(DATA_PATH)
print(f"Loaded {len(df):,} patient records")
df.head()
```

```python
# Descriptive statistics
df.describe()
```

```python
# Outcome distribution
if 'Outcome' in df.columns:
    outcome = df['Outcome'].value_counts()
    print(f"Class distribution:")
    print(f"  Negative: {outcome.get(0, 'N/A')} ({outcome.get(0, 0)/len(df)*100:.1f}%)")
    print(f"  Positive: {outcome.get(1, 'N/A')} ({outcome.get(1, 0)/len(df)*100:.1f}%)")
```

```python
# Feature correlations
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='RdYlBu', center=0, fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```
