```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

plt.style.use('seaborn-v0_8-whitegrid')
DATA_PATH = Path.home() / 'lab' / 'data' / 'CSVs' / 'EPA_SmartLocationDatabase_V3_Jan_2021_Final.csv'
```

```python
# Load dataset
print("Loading EPA Smart Location data...")
df = pd.read_csv(DATA_PATH, low_memory=False)
print(f"Loaded {len(df):,} census block groups")
print(f"Columns: {len(df.columns)}")
```

```python
# Key walkability columns
walkability_cols = ['NatWalkInd', 'D2A_JPHH', 'D2B_E8MIXA', 'D3B', 'D4A']
df[walkability_cols].describe()
```

```python
# National Walkability Index distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(df['NatWalkInd'].dropna(), bins=50, color='seagreen', edgecolor='white')
axes[0].set_title('National Walkability Index Distribution')
axes[0].set_xlabel('Walkability Score')
axes[0].axvline(df['NatWalkInd'].median(), color='red', linestyle='--', label=f"Median: {df['NatWalkInd'].median():.1f}")
axes[0].legend()

# By state
state_walk = df.groupby('STATEFP')['NatWalkInd'].mean().sort_values(ascending=False).head(20)
axes[1].barh(state_walk.index.astype(str), state_walk.values, color='teal')
axes[1].set_title('Average Walkability by State (Top 20)')
axes[1].set_xlabel('Avg Walkability Index')

plt.tight_layout()
plt.show()
```

```python
# Correlation analysis
key_vars = ['NatWalkInd', 'D1C', 'D2A_JPHH', 'D3B', 'D4A', 'TotPop', 'AutoOwn0']
existing_vars = [v for v in key_vars if v in df.columns]

corr = df[existing_vars].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='RdYlGn', center=0, fmt='.2f')
plt.title('Correlation Matrix: Walkability Variables')
plt.tight_layout()
plt.show()
```
