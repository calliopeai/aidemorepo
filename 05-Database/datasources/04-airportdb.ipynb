## 💻 Complex SQL Example

```python
%%calliope run-sql airportdb
-- Airport connectivity analysis
SELECT 
    origin.iata_code as airport,
    origin.name as airport_name,
    origin.city,
    COUNT(DISTINCT f.destination) as destinations_served,
    COUNT(*) as total_outbound_flights,
    COUNT(DISTINCT f.airplane_id) as aircraft_count
FROM airport origin
JOIN flight f ON origin.iata_code = f.origin
GROUP BY origin.iata_code, origin.name, origin.city
HAVING destinations_served > 5
ORDER BY destinations_served DESC, total_outbound_flights DESC
LIMIT 20
```
