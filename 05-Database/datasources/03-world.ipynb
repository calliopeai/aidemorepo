## 💻 Complex SQL Example

```python
%%calliope run-sql world
SELECT 
    co.Continent,
    COUNT(DISTINCT co.Code) as num_countries,
    SUM(co.Population) as total_population,
    AVG(co.LifeExpectancy) as avg_life_expectancy,
    SUM(co.GNP) as total_gnp,
    AVG(co.GNP / co.Population) as avg_gnp_per_capita,
    COUNT(DISTINCT ci.ID) as num_cities
FROM country co
LEFT JOIN city ci ON co.Code = ci.CountryCode
GROUP BY co.Continent
ORDER BY total_population DESC
```
