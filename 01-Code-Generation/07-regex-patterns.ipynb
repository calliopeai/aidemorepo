```python
CLAUDE = "anthropic-chat:claude-sonnet-4-5-20250929"
MODEL = CLAUDE
```

```python
%%ai $MODEL
Generate regex patterns for common validations with Python code:

1. Email addresses (RFC 5322 compliant)
2. Phone numbers (international format)
3. URLs (http/https with optional query params)
4. Credit card numbers (Visa, Mastercard, Amex)
5. IP addresses (IPv4 and IPv6)

Include test cases for each.
```

```python
%%ai $MODEL
Create regex patterns for date/time extraction:

1. ISO 8601 dates (2024-01-15)
2. US format (01/15/2024 or January 15, 2024)
3. Timestamps (2024-01-15T10:30:00Z)
4. Relative dates ("3 days ago", "next week")
5. Duration strings ("2h 30m", "1 day 5 hours")

Use named groups for extraction.
```

```python
%%ai $MODEL
Generate log parsing regex patterns:

Parse these log formats:
1. Apache access log
2. Nginx error log
3. Python logging format
4. JSON structured logs
5. Syslog format

Extract: timestamp, level, source, message
```

```python
%%ai $MODEL
Create code analysis regex patterns:

1. Find function definitions (Python and JavaScript)
2. Extract import statements
3. Match TODO/FIXME comments
4. Find SQL injection vulnerabilities
5. Detect hardcoded secrets/API keys

Use re.MULTILINE and verbose mode.
```

```python
%%ai $MODEL
Generate a regex-based text processor class:

Features:
- Markdown to plain text converter
- HTML tag stripper
- Whitespace normalizer
- PII (Personal Identifiable Information) redactor
- URL and email link extractor

Include type hints and docstrings.
```
