````python
%%ai $MODEL
Refactor to use the Factory pattern:

```python
def create_notification(type, message, recipient):
    if type == 'email':
        return {'type': 'email', 'to': recipient, 'body': message, 'subject': 'Notification'}
    elif type == 'sms':
        return {'type': 'sms', 'phone': recipient, 'text': message[:160]}
    elif type == 'push':
        return {'type': 'push', 'device_id': recipient, 'message': message, 'sound': 'default'}
    elif type == 'slack':
        return {'type': 'slack', 'channel': recipient, 'text': message, 'username': 'NotifyBot'}
```

Create proper classes with inheritance.
````

````python
%%ai $MODEL
Convert this procedural code to object-oriented:

```python
def calculate_employee_salary(employee_data):
    base = employee_data['base_salary']
    bonus = 0
    
    if employee_data['type'] == 'sales':
        bonus = employee_data['sales'] * 0.1
    elif employee_data['type'] == 'developer':
        bonus = 5000 if employee_data['projects_completed'] > 5 else 2000
    elif employee_data['type'] == 'manager':
        bonus = len(employee_data['team']) * 500
    
    tax = (base + bonus) * 0.3
    return base + bonus - tax
```

Use polymorphism for different employee types.
````
