## Learning Objectives
- Generate sorting algorithms with AI
- Create search algorithms
- Implement dynamic programming solutions
- Optimize algorithm complexity

```python
CLAUDE = "anthropic-chat:claude-sonnet-4-5-20250929"
MODEL = CLAUDE
```

```python
%%ai $MODEL
Generate a Python implementation of quicksort with:
- Clear comments explaining each step
- Median-of-three pivot selection
- Handling of duplicate elements
- Include time complexity analysis
```

```python
%%ai $MODEL
Implement binary search with these variations:
1. Standard binary search
2. Find first occurrence
3. Find last occurrence
4. Find insertion point

Include test cases for each.
```

```python
%%ai $MODEL
Create a dynamic programming solution for the knapsack problem:
- 0/1 Knapsack (items can't be divided)
- Include memoization and tabulation approaches
- Track which items are selected
- Demonstrate with example weights and values
```

```python
%%ai $MODEL
Implement Dijkstra's shortest path algorithm:
- Use a priority queue for efficiency
- Return both distances and paths
- Handle disconnected graphs
- Include visualization-ready output
```

```python
%%ai $MODEL
Generate a solution for the longest common subsequence (LCS) problem:
- Dynamic programming approach
- Reconstruct the actual LCS string
- Space-optimized version
- Compare with recursive approach
```
