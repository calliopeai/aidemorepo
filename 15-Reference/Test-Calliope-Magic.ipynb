# 🧪 Test Calliope Magic Commands

This notebook tests the `%%calliope` magic command implementation.

## Setup

The magic command requires:
1. Calliope Data Agent running on port 5000
2. `CALLIOPE_API_BASE_URL` environment variable set
3. The calliope_magic.py extension loaded

## Step 1: Load the Extension

```python
import sys
sys.path.insert(0, '/mnt/efs/users/lmata/lab/aidemorepo')

%load_ext calliope_magic
```

## Step 2: Configure API Endpoint

```python
import os

# Configure Calliope Data Agent API endpoint
API_BASE_URL = 'http://localhost:5000'  # Local development
# API_BASE_URL = 'http://data-agent:5000'  # Docker/JupyterHub

os.environ['CALLIOPE_API_BASE_URL'] = API_BASE_URL
print(f"✅ Calliope API configured: {API_BASE_URL}")
```

## Step 3: Test API Connection

```python
import requests

try:
    response = requests.get('http://localhost:5000/api/datasources')
    datasources = response.json()['datasources']
    print(f"✅ Connected! Found {len(datasources)} datasources:")
    for ds in datasources:
        print(f"  • {ds['id']}: {ds['name']} ({ds['dialect']})")
except Exception as e:
    print(f"❌ Connection failed: {e}")
```

## Test 2: ask-sql Command

```python
%%calliope run-sql world --format table
SELECT Name, Continent, Population 
FROM country 
ORDER BY Population DESC 
LIMIT 10
```

## Summary

If all tests pass, the `%%calliope` magic command is working correctly!

### Troubleshooting

**Error: "Could not connect to Calliope API"**
- Check if Data Agent is running: `docker-compose ps data-agent`
- Verify API is responding: `curl http://localhost:5000/api/datasources`

**Error: "Missing datasource_id"**
- Make sure to specify a datasource after the subcommand: `%%calliope ask-sql sakila`

**Error: "Unknown subcommand"**
- Valid subcommands are: `ask-sql`, `run-sql`, `generate-sql`
