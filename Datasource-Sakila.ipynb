## 💻 Step 11: Direct SQL Queries

For developers who want to write custom SQL:

```python
%%calliope run-sql sakila
SELECT 
    c.name AS category,
    COUNT(DISTINCT f.film_id) AS film_count,
    AVG(f.rental_rate) AS avg_rental_rate,
    AVG(f.length) AS avg_length
FROM category c
JOIN film_category fc ON c.category_id = fc.category_id
JOIN film f ON fc.film_id = f.film_id
GROUP BY c.category_id, c.name
ORDER BY film_count DESC
```

```python
%%calliope run-sql sakila
SELECT 
    f.title,
    f.rating,
    COUNT(r.rental_id) AS rental_count,
    SUM(p.amount) AS total_revenue,
    COUNT(DISTINCT i.inventory_id) AS inventory_copies,
    ROUND(SUM(p.amount) / COUNT(DISTINCT i.inventory_id), 2) AS revenue_per_copy
FROM film f
JOIN inventory i ON f.film_id = i.film_id
JOIN rental r ON i.inventory_id = r.inventory_id
JOIN payment p ON r.rental_id = p.rental_id
GROUP BY f.film_id, f.title, f.rating
ORDER BY revenue_per_copy DESC
LIMIT 20
```
