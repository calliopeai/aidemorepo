## 🗄️ Step 3: Add a Database Connection

Let's connect to a sample SQLite database. SQLite is perfect for demos as it requires no server setup.

````python
# First, let's create a sample SQLite database with some demo data
import sqlite3
import pandas as pd

# Create connection
conn = sqlite3.connect('sample_sales.db')

# Create sample sales data
sales_data = pd.DataFrame({
    'order_id': range(1, 11),
    'customer_name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank', 'Grace', 'Henry', 'Iris', 'Jack'],
    'product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Laptop', 'Mouse'],
    'quantity': [1, 2, 1, 1, 2, 3, 1, 2, 1, 5],
    'price': [1200, 25, 75, 300, 1200, 25, 75, 300, 1200, 25],
    'order_date': ['2024-01-15', '2024-01-16', '2024-01-17', '2024-01-18', '2024-01-19', 
                   '2024-01-20', '2024-01-21', '2024-01-22', '2024-01-23', '2024-01-24']
})

# Calculate total
sales_data['total'] = sales_data['quantity'] * sales_data['price']

# Write to database
sales_data.to_sql('orders', conn, if_exists='replace', index=False)

conn.close()
print("✅ Sample database created: sample_sales.db")
print("\nSample data:")
print(sales_data.head())
````

## 💡 Power Tips

### Tip 1: Chain Multiple Queries
```python
# First query
%%calliope ask-sql sales_demo
What are total sales by product?

# Follow-up using results
%%calliope ask-sql sales_demo
Which products have above-average sales?
```

### Tip 2: Combine with Pandas
```python
import pandas as pd

# Get results from Calliope
%%calliope ask-sql sales_demo --format pandas
SELECT * FROM orders

# Then process with pandas
df = _  # Get last output
df['profit_margin'] = (df['price'] - df['cost']) / df['price']
df.head()
```

### Tip 3: Use Different Models for Different Tasks
- **Complex analytical SQL**: `--sql-model anthropic`
- **Fast simple queries**: `--sql-model gemini`  
- **General purpose**: `--sql-model openai`

### Tip 4: Generate SQL First (Debug Mode)
```python
%%calliope generate-sql sales_demo
Calculate year-over-year growth by product

# Review the SQL, then execute if good:
%%calliope run-sql sales_demo
<paste generated SQL here>
```
