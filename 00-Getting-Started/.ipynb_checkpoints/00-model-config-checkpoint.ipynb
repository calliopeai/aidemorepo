# Model Configuration & Aliases

**Purpose:** Central configuration for AI models used across all demo notebooks  
**Last Updated:** 2025-12

---

## Why This Matters

AI model names change frequently. By centralizing configuration:
- Update once, use everywhere
- Easy switching between providers
- Consistent aliases across all notebooks

---

## Quick Start

Copy this configuration cell to the top of any notebook:

````python
# =============================================================================
# MODEL ALIASES CONFIGURATION - December 2025
# =============================================================================
# Update these when models are deprecated or new ones are released
# Run: %calliope list-models  to see all available options
# =============================================================================

# ANTHROPIC CLAUDE (Best for complex SQL and analysis)
CLAUDE = "anthropic-chat:claude-sonnet-4-5-20250929"          # Default
CLAUDE_FAST = "anthropic-chat:claude-haiku-4-5-20251001"      # Fast/cheap
CLAUDE_BEST = "anthropic-chat:claude-opus-4-1-20250805"       # Most capable

# OPENAI GPT
GPT = "openai-chat:gpt-5"                   # Default GPT
GPT_FAST = "openai-chat:gpt-5-mini"         # Fast/cheap
GPT4O = "openai-chat:gpt-4o"                # GPT-4o

# GOOGLE GEMINI
GEMINI = "gemini:gemini-2.5-flash"          # Fast
GEMINI_BEST = "gemini:gemini-2.5-pro"       # Most capable

# MISTRAL
MISTRAL = "mistralai:mistral-large-latest"  # Default Mistral
CODESTRAL = "mistralai:codestral-latest"    # Code-specialized

# COHERE
COHERE = "cohere:command-a"                 # Latest Command

# AWS BEDROCK (No API keys - uses AWS credentials)
BEDROCK_CLAUDE = "bedrock-chat:anthropic.claude-3-5-sonnet-20241022-v2:0"
BEDROCK_LLAMA = "bedrock-chat:meta.llama3-1-405b-instruct-v1:0"

# =============================================================================
# RECOMMENDED DEFAULTS FOR DIFFERENT TASKS
# =============================================================================
SQL_MODEL = CLAUDE          # Complex SQL generation
ANALYSIS_MODEL = CLAUDE     # Data analysis and insights
FAST_MODEL = GEMINI         # Quick queries, simple tasks
CODE_MODEL = CODESTRAL      # Code generation

print("Model aliases configured!")
print(f"SQL_MODEL: {SQL_MODEL}")
print(f"ANALYSIS_MODEL: {ANALYSIS_MODEL}")
````

## Available Models

Run this to see all available models in your environment:

````python
%calliope list-models
````

## Usage Examples

### With Calliope Database Queries

```python
# Use the SQL_MODEL alias for complex queries
%calliope -d sakila --sql-model $SQL_MODEL "Show revenue by category"

# Use FAST_MODEL for simple queries
%calliope -d sakila --sql-model $FAST_MODEL "How many films?"

# Get AI analysis of results
%calliope -d sakila --sql-model $SQL_MODEL --to-ai $ANALYSIS_MODEL "Analyze customer trends"
```

### With Jupyter AI Magic

```python
%%ai $CLAUDE
Explain the difference between INNER and OUTER joins

%%ai $GPT -f code
Write a Python function to calculate moving average
```

---

## Model Selection Guide

| Task | Recommended Model | Why |
|------|------------------|-----|
| Complex SQL | CLAUDE | Best at multi-table joins, window functions |
| Simple queries | GEMINI | Fastest, good for basic SQL |
| Data analysis | CLAUDE or GPT | Deep analytical reasoning |
| Code generation | CODESTRAL | Specialized for code |
| Quick answers | GEMINI | Lowest latency |
| AWS environment | BEDROCK_CLAUDE | No external API keys needed |

---

## Updating Models

When models are deprecated:

1. Run `%calliope list-models` to see current options
2. Update the aliases in the configuration cell
3. Re-run the configuration cell
4. All notebooks using these aliases will use the new models
