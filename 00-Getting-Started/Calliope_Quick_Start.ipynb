## Quick Reference

### Always Clean Column Names First:
```python
df.columns = df.columns.str.strip()  # Remove whitespace
df.columns = df.columns.str.replace(' ', '_')  # Replace spaces
df.columns = df.columns.str.lower()  # Lowercase (optional)
```

### Syntax:
```python
%%calliope ask model_name --format code
Your natural language request here
```

### Popular Models:
- **Fast**: `gemini:gemini-2.5-flash`, `claude-haiku-4-5-20251001`, `gpt-4o-mini`
- **Balanced**: `gpto`, `gpt-4o`, `claude-sonnet-4-5-20250929`, `gemini:gemini-2.5-pro`
- **Powerful**: `gpt-5`, `claude-opus-4-20250514`, `mistralai:magistral-medium`

### Tips:
1. **Always clean column names** in your prompt to help the AI
2. Use `--format code` for executable code, `--format text` for explanations
3. Tell AI to "store as variable_name" to save results
4. Reference previous dataframes by name in next cells
5. Sample large datasets for faster processing
6. Request specific plot types and layouts
