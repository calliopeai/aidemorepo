## 🗄️ Setup: Create a Rich Sample Database

```python
import sqlite3
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Create connection
conn = sqlite3.connect('ecommerce_analytics.db')

# Create customers table
customers = pd.DataFrame({
    'customer_id': range(1, 101),
    'customer_name': [f'Customer_{i}' for i in range(1, 101)],
    'city': np.random.choice(['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'], 100),
    'state': np.random.choice(['NY', 'CA', 'IL', 'TX', 'AZ'], 100),
    'signup_date': [(datetime(2023, 1, 1) + timedelta(days=np.random.randint(0, 365))).strftime('%Y-%m-%d') for _ in range(100)]
})

# Create products table
products = pd.DataFrame({
    'product_id': range(1, 21),
    'product_name': [f'Product_{i}' for i in range(1, 21)],
    'category': np.random.choice(['Electronics', 'Clothing', 'Home', 'Sports', 'Books'], 20),
    'price': np.random.uniform(10, 500, 20).round(2)
})

# Create orders table (500 orders)
np.random.seed(42)
orders = pd.DataFrame({
    'order_id': range(1, 501),
    'customer_id': np.random.randint(1, 101, 500),
    'product_id': np.random.randint(1, 21, 500),
    'quantity': np.random.randint(1, 5, 500),
    'order_date': [(datetime(2024, 1, 1) + timedelta(days=np.random.randint(0, 90))).strftime('%Y-%m-%d') for _ in range(500)],
    'status': np.random.choice(['completed', 'pending', 'cancelled'], 500, p=[0.85, 0.10, 0.05])
})

# Write to database
customers.to_sql('customers', conn, if_exists='replace', index=False)
products.to_sql('products', conn, if_exists='replace', index=False)
orders.to_sql('orders', conn, if_exists='replace', index=False)

conn.close()
print("✅ E-commerce analytics database created!")
print(f"\n📊 Database contents:")
print(f"- Customers: {len(customers)} records")
print(f"- Products: {len(products)} records")
print(f"- Orders: {len(orders)} records")
```

```python
%%calliope add-database
{
    "datasource_id": "ecommerce",
    "name": "E-Commerce Analytics",
    "dialect": "sqlite",
    "connection_details": {
        "database": "ecommerce_analytics.db"
    }
}
```
