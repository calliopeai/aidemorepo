```python
CLAUDE = "anthropic-chat:claude-sonnet-4-5-20250929"
MODEL = CLAUDE
```

```python
%%ai $MODEL
Create a complete linear regression pipeline:
1. Generate synthetic housing data (size, bedrooms, age, price)
2. Exploratory data analysis with visualizations
3. Train/test split
4. Fit linear regression model
5. Evaluate with R², MSE, MAE
6. Visualize predictions vs actuals
```

```python
%%ai $MODEL
Compare regularization techniques:
- Lasso (L1) regression
- Ridge (L2) regression
- ElasticNet

Use cross-validation to find optimal alpha.
Compare feature coefficients and model performance.
Explain when to use each technique.
```

```python
%%ai $MODEL
Implement polynomial regression:
1. Create non-linear synthetic data
2. Compare linear vs polynomial fits (degree 2, 3, 5)
3. Demonstrate overfitting with high degree
4. Use cross-validation to select optimal degree
5. Plot learning curves
```

```python
%%ai $MODEL
Build a gradient boosting regressor:
- Use XGBoost or LightGBM
- Feature importance analysis
- Hyperparameter tuning with GridSearchCV
- SHAP values for interpretability
- Compare with linear regression baseline
```

```python
%%ai $MODEL
Handle multicollinearity in regression:
1. Generate correlated features
2. Detect with correlation matrix and VIF
3. Show impact on coefficient stability
4. Apply PCA to reduce collinearity
5. Compare model stability before/after
```
