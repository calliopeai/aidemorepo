```python
import requests
import pandas as pd
import plotly.express as px
import dash
from dash import dcc, html
from dash.dependencies import Input, Output

# Fetch live COVID-19 data
url = 'https://covid.ourworldindata.org/data/owid-covid-data.csv'
data = pd.read_csv(url)

# Data Cleaning
data = data[['location', 'date', 'total_cases', 'new_cases', 'total_deaths', 'new_deaths']]
data['date'] = pd.to_datetime(data['date'])

# Dash app setup
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("COVID-19 Data Tracker"),
    dcc.Dropdown(
        id='country-dropdown',
        options=[
            {'label': country, 'value': country} for country in data['location'].unique()
        ],
        value='World',
        multi=False
    ),
    dcc.Graph(id='covid-graph')
])

@app.callback(
    Output('covid-graph', 'figure'),
    [Input('country-dropdown', 'value')]
)
def update_graph(selected_country):
    filtered_data = data[data['location'] == selected_country]
    
    fig = px.line(filtered_data, x='date', y='total_cases', 
                  title=f'COVID-19 Total Cases in {selected_country}', 
                  labels={'total_cases': 'Total Cases', 'date': 'Date'})
    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```
