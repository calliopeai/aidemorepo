```python
# Final Combined Web Service Script
#
# This script launches two independent web services on different ports.
#
# Dependencies:
# pip install Flask numpy plotly
#
# How to Run:
# 1. Make sure you have Python and the required libraries installed.
# 2. Save this code as a single Python file (e.g., cool_services.py).
# 3. Run from your terminal: python cool_services.py
#
# Access the services:
# - Service 1 (Neon 3D Background): http://127.0.0.1:9191
# - Service 2 (Math Visualization): http://127.0.0.1:9200

import threading
import time
import logging

# Suppress the default Flask startup messages to keep the output clean.
log = logging.getLogger('werkzeug')
log.setLevel(logging.ERROR)

# --- Service 1: Port 9191 - Neon 3D Shapes Background ---

from flask import Flask, Response

# The entire web page for service 1 is self-contained in this string.
# It includes HTML, CSS, and JavaScript using modern ES6 modules from a CDN.
HTML_9191 = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Dark Service [9191]</title>
    <style>
        body {
            margin: 0;
            padding: 0;
            background-color: #000000;
            color: #00ff7f; /* Neon Green */
            font-family: 'Courier New', Courier, monospace;
            overflow: hidden; /* Hide scrollbars */
            text-align: center;
        }
        canvas {
            display: block;
            position: fixed;
            top: 0;
            left: 0;
            z-index: -1; /* Place canvas behind other content */
        }
        .content {
            position: relative;
            z-index: 1;
            padding-top: 35vh;
            text-shadow: 0 0 5px #00ff7f, 0 0 10px #00ff7f, 0 0 20px #00ff7f;
        }
        h1 {
            font-size: 4rem;
            margin: 0;
        }
        p {
            font-size: 1.5rem;
        }
    </style>
</head>
<body>
    <div class="content">
        <h1>SERVICE 9191</h1>
        <p>GEOMETRIC NEON FIELD</p>
    </div>

    <!-- The 3D magic happens here -->
    <script type="module">
        // Import necessary modules from Three.js CDN
        import * as THREE from 'https://cdn.skypack.dev/three@0.136.0';
        import { EffectComposer } from 'https://cdn.skypack.dev/three@0.136.0/examples/jsm/postprocessing/EffectComposer.js';
        import { RenderPass } from 'https://cdn.skypack.dev/three@0.136.0/examples/jsm/postprocessing/RenderPass.js';
        import { UnrealBloomPass } from 'https://cdn.skypack.dev/three@0.136.0/examples/jsm/postprocessing/UnrealBloomPass.js';

        // Scene setup
        const scene = new THREE.Scene();
        const camera = new THREE.PerspectiveCamera(75, window.innerWidth / window.innerHeight, 0.1, 1000);
        camera.position.z = 50;

        const renderer = new THREE.WebGLRenderer({ antialias: true });
        renderer.setSize(window.innerWidth, window.innerHeight);
        renderer.setPixelRatio(window.devicePixelRatio);
        renderer.toneMapping = THREE.ReinhardToneMapping;
        document.body.appendChild(renderer.domElement);

        // Post-processing for the "neon" bloom effect
        const renderScene = new RenderPass(scene, camera);
        const bloomPass = new UnrealBloomPass(new THREE.Vector2(window.innerWidth, window.innerHeight), 1.5, 0.4, 0.85);
        bloomPass.threshold = 0;
        bloomPass.strength = 1.8; // Intensity of the glow
        bloomPass.radius = 0.5;

        const composer = new EffectComposer(renderer);
        composer.addPass(renderScene);
        composer.addPass(bloomPass);

        // Procedural Generation of Shapes
        const shapesGroup = new THREE.Group();
        const NUM_SHAPES = 50;
        const SPREAD_RADIUS = 100;

        const geometries = [
            new THREE.IcosahedronGeometry(2, 0),
            new THREE.DodecahedronGeometry(2, 0),
            new THREE.TorusKnotGeometry(2, 0.5, 100, 16),
            new THREE.TorusGeometry(2, 0.8, 16, 100),
            new THREE.OctahedronGeometry(2, 0)
        ];

        const colors = [0x00ff7f, 0xff00ff, 0x00ffff, 0xffff00];

        for (let i = 0; i < NUM_SHAPES; i++) {
            const geometry = geometries[Math.floor(Math.random() * geometries.length)];
            const material = new THREE.MeshBasicMaterial({
                color: colors[Math.floor(Math.random() * colors.length)],
                wireframe: true
            });
            const shape = new THREE.Mesh(geometry, material);

            // Position shapes randomly in a spherical volume
            const phi = Math.random() * Math.PI * 2;
            const theta = Math.acos((Math.random() * 2) - 1);
            const radius = Math.random() * SPREAD_RADIUS;
            
            shape.position.x = radius * Math.sin(theta) * Math.cos(phi);
            shape.position.y = radius * Math.sin(theta) * Math.sin(phi);
            shape.position.z = radius * Math.cos(theta);
            
            shape.rotation.set(Math.random() * Math.PI, Math.random() * Math.PI, Math.random() * Math.PI);
            
            shapesGroup.add(shape);
        }
        scene.add(shapesGroup);

        // Handle window resizing
        window.addEventListener('resize', () => {
            const width = window.innerWidth;
            const height = window.innerHeight;
            camera.aspect = width / height;
            camera.updateProjectionMatrix();
            renderer.setSize(width, height);
            composer.setSize(width, height);
        });

        // Animation loop
        function animate() {
            requestAnimationFrame(animate);

            // Slowly rotate the entire group of shapes
            shapesGroup.rotation.x += 0.0005;
            shapesGroup.rotation.y += 0.0007;

            // Use the composer to render the scene with the bloom effect
            composer.render();
        }

        animate();
    </script>
</body>
</html>
"""

app_9191 = Flask(__name__)

@app_9191.route('/')
def home_9191():
    return Response(HTML_9191, mimetype='text/html')

def run_server_9191():
    print("-> Starting darkmode webservice on http://127.0.0.1:9191")
    app_9191.run(host='0.0.0.0', port=9191)

# --- Service 2: Port 9200 - Jupyter-like Math Visualization ---

from flask import render_template_string
import numpy as np
import plotly.graph_objects as go

# HTML template for the math visualization page.
# The {{ plot_div|safe }} is a placeholder where Flask will inject the plot.
HTML_9200_TEMPLATE = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Math Viz [9200]</title>
    <style>
        body {
            background-color: #111111;
            color: #eeeeee;
            font-family: -apple-system, BlinkMacSystemFont, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif;
            margin: 0;
            padding: 20px;
            display: flex;
            flex-direction: column;
            align-items: center;
        }
        h1 {
            color: #00aaff;
            text-shadow: 0 0 5px #00aaff;
        }
        .plot-container {
            width: 95vw;
            height: 85vh;
        }
    </style>
</head>
<body>
    <h1>SERVICE 9200</h1>
    <p>Interactive 3D Sinc Function Visualization</p>
    <div class="plot-container">
        {{ plot_div|safe }}
    </div>
</body>
</html>
"""

app_9200 = Flask(__name__)

@app_9200.route('/')
def home_9200():
    # Generate data for a 3D Sinc function plot
    x = np.arange(-15, 15, 0.5)
    y = np.arange(-15, 15, 0.5)
    xx, yy = np.meshgrid(x, y)
    R = np.sqrt(xx**2 + yy**2)
    # Add a small epsilon to avoid division by zero at the center
    zz = np.sin(R) / (R + 1e-8)

    # Create a Plotly 3D Surface plot
    fig = go.Figure(data=[go.Surface(z=zz, x=xx, y=yy, colorscale='Cividis_r', showscale=False)])

    # Configure the layout for a cool, dark, Jupyter-like appearance
    fig.update_layout(
        title={
            'text': 'f(x, y) = sin(sqrt(x² + y²)) / sqrt(x² + y²)',
            'y':0.9,
            'x':0.5,
            'xanchor': 'center',
            'yanchor': 'top'
        },
        autosize=True,
        template='plotly_dark',
        scene=dict(
            xaxis_title='X-axis',
            yaxis_title='Y-axis',
            zaxis_title='Z-axis',
            xaxis=dict(showbackground=False, showticklabels=False, title=''),
            yaxis=dict(showbackground=False, showticklabels=False, title=''),
            zaxis=dict(showbackground=False, showticklabels=False, title=''),
            camera_eye=dict(x=1.8, y=1.8, z=1.4) # Set a nice initial camera angle
        ),
        margin=dict(l=0, r=0, b=0, t=40) # Minimize margins
    )

    # Convert the figure to an HTML div
    plot_div = fig.to_html(full_html=False, include_plotlyjs='cdn')

    # Render the template, injecting the plot
    return render_template_string(HTML_9200_TEMPLATE, plot_div=plot_div)

def run_server_9200():
    print("-> Starting math visualization service on http://127.0.0.1:9200")
    app_9200.run(host='0.0.0.0', port=9200)


# --- Main execution block to run both servers in parallel ---

if __name__ == '__main__':
    # Create threads for each server
    thread_9191 = threading.Thread(target=run_server_9191)
    thread_9200 = threading.Thread(target=run_server_9200)

    # Set threads as daemons so they will exit when the main thread exits
    thread_9191.daemon = True
    thread_9200.daemon = True

    # Start the servers
    thread_9191.start()
    thread_9200.start()

    print("\n-----------------------------------------------------")
    print("Two web services have been launched in the background.")
    print("Press Ctrl+C to stop the servers.")
    print("-----------------------------------------------------\n")

    # Keep the main thread alive to allow daemon threads to run
    try:
        while True:
            time.sleep(1)
    except KeyboardInterrupt:
        print("\nShutting down servers...")
```
