## 🗄️ Step 3: Add a Database Connection

Let's connect to a sample SQLite database. SQLite is perfect for demos as it requires no server setup.

````python
# First, let's create a sample SQLite database with some demo data
import sqlite3
import pandas as pd

# Create connection
conn = sqlite3.connect('sample_sales.db')

# Create sample sales data
sales_data = pd.DataFrame({
    'order_id': range(1, 11),
    'customer_name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank', 'Grace', 'Henry', 'Iris', 'Jack'],
    'product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Laptop', 'Mouse'],
    'quantity': [1, 2, 1, 1, 2, 3, 1, 2, 1, 5],
    'price': [1200, 25, 75, 300, 1200, 25, 75, 300, 1200, 25],
    'order_date': ['2024-01-15', '2024-01-16', '2024-01-17', '2024-01-18', '2024-01-19', 
                   '2024-01-20', '2024-01-21', '2024-01-22', '2024-01-23', '2024-01-24']
})

# Calculate total
sales_data['total'] = sales_data['quantity'] * sales_data['price']

# Write to database
sales_data.to_sql('orders', conn, if_exists='replace', index=False)

conn.close()
print("✅ Sample database created: sample_sales.db")
print("\nSample data:")
print(sales_data.head())
````

````python
%%calliope add-database
{
    "datasource_id": "sales_demo",
    "name": "Sales Demo Database",
    "dialect": "sqlite",
    "connection_details": {
        "database": "sample_sales.db"
    }
}
````

## 💬 Step 4: Ask Your First Natural Language Question

Now let's query the database using natural language. Calliope will:
1. Convert your question to SQL
2. Execute the query
3. Return results with visualizations

````python
%%calliope ask-sql sales_demo
What are the total sales by product?
````

````python
%%calliope ask-sql sales_demo
Who are our top 3 customers by total revenue?
````

````python
%%calliope ask-sql sales_demo
Show me daily sales trends
````

## 🎯 Step 5: Choose Different AI Models

You can specify which AI model to use for SQL generation:

````python
%%calliope ask-sql sales_demo --sql-model openai
What's the average order value?
````

````python
%%calliope ask-sql sales_demo --sql-model anthropic
Calculate the total revenue and number of orders
````
