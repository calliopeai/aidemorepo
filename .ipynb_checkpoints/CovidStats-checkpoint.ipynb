```python
import requests
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import ipywidgets as widgets
from IPython.display import display

# Step 1: Fetch live COVID-19 data
def fetch_covid_data():
    url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
    response = requests.get(url)
    with open('covid_data.csv', 'wb') as file:
        file.write(response.content)
    return pd.read_csv('covid_data.csv')

# Step 2: Data Cleaning
def clean_data(df):
    df['date'] = pd.to_datetime(df['date'])
    df = df.dropna(subset=['location', 'total_cases'])
    return df

# Step 3: Data Analysis
def analyze_data(df):
    latest_data = df[df['date'] == df['date'].max()]
    cases_by_country = latest_data.groupby('location')['total_cases'].sum().reset_index()
    return cases_by_country

# Step 4: Data Visualization
def plot_data(df):
    fig = px.choropleth(df, locations="location", locationmode='country names',
                        color="total_cases", hover_name="location",
                        color_continuous_scale=px.colors.sequential.Plasma,
                        title="COVID-19 Total Cases by Country")
    return fig

# Step 5: Interactive Dashboard
def create_dashboard():
    # Fetch and clean data
    df = fetch_covid_data()
    df_cleaned = clean_data(df)
    
    # Analyze data
    cases_by_country = analyze_data(df_cleaned)
    
    # Create Plot
    fig = plot_data(cases_by_country)
    
    # Create widgets for interaction
    country_dropdown = widgets.Dropdown(
        options=cases_by_country['location'].unique(),
        description='Country:',
    )
    
    def update_plot(change):
        filtered_df = df_cleaned[df_cleaned['location'] == change['new']]
        fig = go.Figure()
        fig.add_trace(go.Scatter(x=filtered_df['date'], y=filtered_df['total_cases'], mode='lines', name='Total Cases'))
        fig.update_layout(title=f"COVID-19 Total Cases Over Time in {change['new']}", xaxis_title="Date", yaxis_title="Total Cases")
        fig.show()
    
    country_dropdown.observe(update_plot, names='value')
    
    display(country_dropdown)

# Run the dashboard
create_dashboard()
```

```python
import requests
import pandas as pd
import plotly.express as px
import dash
from dash import dcc, html
from dash.dependencies import Input, Output

# Fetch live COVID-19 data
url = 'https://covid.ourworldindata.org/data/owid-covid-data.csv'
data = pd.read_csv(url, nrows=1000)

# Data Cleaning
data = data[['location', 'date', 'total_cases', 'new_cases', 'total_deaths', 'new_deaths']]
data['date'] = pd.to_datetime(data['date'])

# Dash app setup
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("COVID-19 Data Tracker"),
    dcc.Dropdown(
        id='country-dropdown',
        options=[
            {'label': country, 'value': country} for country in data['location'].unique()
        ],
        value='World',
        multi=False
    ),
    dcc.Graph(id='covid-graph')
])

@app.callback(
    Output('covid-graph', 'figure'),
    [Input('country-dropdown', 'value')]
)
def update_graph(selected_country):
    filtered_data = data[data['location'] == selected_country]
    
    fig = px.line(filtered_data, x='date', y='total_cases', 
                  title=f'COVID-19 Total Cases in {selected_country}', 
                  labels={'total_cases': 'Total Cases', 'date': 'Date'})
    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```
