```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def uneven_sin(x):
    return np.sin(x) + 0.5 * np.sin(3 * x)

# Create an array of x values
x = np.linspace(0, 2 * np.pi, 1000)
# Calculate the y values for the uneven sinusoidal function
y = uneven_sin(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Uneven Sinusoidal Curve')
plt.fill_between(x, y, where=(y > 0), color='blue', alpha=0.3, label='Area under curve')

# Add labels and title
plt.title('Shaded Area Under an Uneven Sinusoidal Curve')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.8)
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sinusoidal function
def f(x):
    return np.sin(x) + 0.5 * np.sin(2*x)

# Define the x-values
x = np.linspace(0, 2*np.pi, 100)

# Calculate the y-values
y = f(x)

# Shade the area under the curve
plt.fill_between(x, y, color='blue', alpha=0.2)

# Plot the sinusoidal function
plt.plot(x, y, color='black')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x) + 0.5 * np.random.rand(100)

fig, ax = plt.subplots()
ax.plot(x, y, '-')
ax.fill_between(x, 0, y)

plt.show()
```
