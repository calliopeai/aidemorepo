```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

sigma = 10
beta = 8/3
rho = 28

def lorenz(x, y, z, sigma=sigma, beta=beta, rho=rho):
    x_dot = sigma * (y - x)
    y_dot = x * (rho - z) - y
    z_dot = x * y - beta * z
    return x_dot, y_dot, z_dot

dt = 0.01
num_steps = 10000

xs = np.empty(num_steps + 1)
ys = np.empty(num_steps + 1)
zs = np.empty(num_steps + 1)

xs[0], ys[0], zs[0] = (0., 1., 1.05)

for i in range(num_steps):
    x_dot, y_dot, z_dot = lorenz(xs[i], ys[i], zs[i])
    xs[i + 1] = xs[i] + (x_dot * dt)
    ys[i + 1] = ys[i] + (y_dot * dt)
    zs[i + 1] = zs[i] + (z_dot * dt)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs, ys, zs, lw=0.5)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz Attractor")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

import math

fig, ax = plt.subplots(2, 4, figsize=(10, 6), sharex=True, sharey=True)

# Initial conditions (2 dimensions)
x_0 = np.linspace(-1, 1, num=1000)
y_0 = np.linspace(-1, 1, num=1000)[::-1]
X, Y = np.meshgrid(x_0, y_0)

# Time steps
t_steps = 100

# Parameters
sigma = 10
beta = 8 / 3
theta = math.pi * 5

# Generate the dynamic system values
t = np.linspace(0, 5, t_steps)
x = np.zeros_like(t)
y = np.zeros_like(t)

for i in range(t_steps):
    x[i] = sigma * np.cos(theta * t[i]) * np.sin(beta * x[i - 1]) + y[i - 1]
    y[i] = sigma * np.sin(theta * t[i]) * np.cos(beta * y[i - 1]) - x[i - 1]

ax[0, 0].contour(X, Y, x, 50, cmap='Greys')
ax[0, 0].set_title('Time = 0')

for i in range(1, 5):
    ax[0, i].contour(X, Y, x[i], 50, cmap='Greys')
    ax[0, i].set_title(f'Time = {i * 0.2}')

for i in range(1, 4):
    ax[1, i].contour(X, Y, y, 50, cmap='Greys')
    ax[1, i].set_title(f'Time = {(i - 1) * 0.2 + 0.4}')

for axi in fig.axes:
    axi.set_xlim(-1, 1)
    axi.set_ylim(-1, 1)
    axi.set_aspect('equal', 'box')

fig.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

import toronto_edu as te

# Generate the Lorenz attractor data
np.random.seed(0)
t = np.linspace(0, 10, 2001)
y = np.linspace(-10, 10, 2001)
x = np.zeros_like(y)
z = np.zeros_like(y)

for i in range(2001):
    x[i], z[i] = np.random.normal(0, 10, 2)
    y[i] = (x[i - 1] + z[i - 1] * np.sin(y[i - 1])) * 0.1 + np.random.normal(0, 0.1, 1)

# Plot the Lorenz attractor
plt.figure(figsize=(10, 8))
plt.subplot(1, 2, 1)
plt.title('Lorenz Attractor')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.scatter(x, y)

plt.subplot(1, 2, 2)
plt.title('Phase Diagram')
plt.xlabel('x')
plt.ylabel('z')
plt.grid(True)
plt.scatter(x, z)

plt.tight_layout()
plt.show()
```
