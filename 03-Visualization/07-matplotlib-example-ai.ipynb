```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some cool data
np.random.seed(42)
x = np.linspace(0, 10, 500)
y1 = np.sin(x) + np.random.normal(0, 0.1, 500)
y2 = np.cos(x) + np.random.normal(0, 0.1, 500)
y3 = np.sin(x * 2) * 0.5 + np.random.normal(0, 0.05, 500)

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(12, 8))

# Plot the data with different styles
ax.plot(x, y1, label='Sine Wave (noisy)', color='dodgerblue', linewidth=2, alpha=0.8)
ax.plot(x, y2, label='Cosine Wave (noisy)', color='orangered', linestyle='--', linewidth=2, alpha=0.8)
ax.plot(x, y3, label='Deeper Sine Wave', color='forestgreen', linewidth=1.5, alpha=0.7, marker='.', markersize=3, markevery=20)

# Add some flair
ax.set_title('A Cool Matplotlib Visualization', fontsize=20, fontweight='bold', color='dimgray')
ax.set_xlabel('X-axis', fontsize=14, color='gray')
ax.set_ylabel('Y-axis', fontsize=14, color='gray')
ax.grid(True, linestyle=':', alpha=0.6, color='lightgray')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('gray')
ax.spines['bottom'].set_color('gray')
ax.tick_params(axis='both', which='major', labelsize=12, colors='dimgray')

# Add a legend
legend = ax.legend(loc='upper right', fontsize=12, frameon=True, shadow=True, facecolor='whitesmoke', edgecolor='gray')
for lh in legend.legendHandles:
    lh.set_alpha(1) # Ensure legend handles are fully opaque

# Add some annotations
ax.annotate('Peak!', xy=(np.pi/2, 1.1), xytext=(np.pi/2 + 1, 1.3),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5, headwidth=8),
            fontsize=12, color='black')

ax.annotate('Trough', xy=(3*np.pi/2, -1.1), xytext=(3*np.pi/2 - 1.5, -1.5),
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2", color='purple', lw=2),
            fontsize=12, color='purple')

# Improve layout
plt.tight_layout()

# Display the plot
plt.show()
```
