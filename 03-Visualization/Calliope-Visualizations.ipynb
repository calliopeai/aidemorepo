## 🗄️ Setup: Create Rich Dataset

```python
import sqlite3
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

conn = sqlite3.connect('retail_analytics.db')

# Generate realistic sales data
np.random.seed(42)
dates = [(datetime(2024, 1, 1) + timedelta(days=i)).strftime('%Y-%m-%d') for i in range(90)]

sales_data = []
for date in dates:
    for _ in range(np.random.randint(10, 30)):
        sales_data.append({
            'sale_date': date,
            'store_id': np.random.randint(1, 6),
            'category': np.random.choice(['Electronics', 'Clothing', 'Food', 'Home', 'Sports']),
            'product': np.random.choice([f'Product_{i}' for i in range(1, 31)]),
            'quantity': np.random.randint(1, 10),
            'unit_price': round(np.random.uniform(5, 200), 2),
            'discount': round(np.random.choice([0, 0.05, 0.10, 0.15, 0.20], p=[0.6, 0.15, 0.15, 0.05, 0.05]), 2)
        })

sales = pd.DataFrame(sales_data)
sales['revenue'] = sales['quantity'] * sales['unit_price'] * (1 - sales['discount'])

# Store information
stores = pd.DataFrame({
    'store_id': range(1, 6),
    'store_name': ['Downtown', 'Mall', 'Airport', 'Suburb', 'Online'],
    'city': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Virtual'],
    'region': ['East', 'West', 'Central', 'South', 'Online']
})

# Customer segments
segments = pd.DataFrame({
    'segment_id': range(1, 5),
    'segment_name': ['Premium', 'Regular', 'Budget', 'Wholesale'],
    'avg_purchase': [250, 100, 50, 500]
})

sales.to_sql('sales', conn, if_exists='replace', index=False)
stores.to_sql('stores', conn, if_exists='replace', index=False)
segments.to_sql('segments', conn, if_exists='replace', index=False)

conn.close()
print(f"✅ Created retail database with {len(sales)} sales records!")
```

```python
%%calliope add-database
{
    "datasource_id": "retail",
    "name": "Retail Analytics",
    "dialect": "sqlite",
    "connection_details": {
        "database": "retail_analytics.db"
    }
}
```

## 💡 Visualization Tips

### Calliope Automatically Chooses:
- **Line charts** for time-series data
- **Bar charts** for categorical comparisons
- **Pie charts** for proportions and percentages
- **Scatter plots** for correlations
- **Heatmaps** for multi-dimensional data
- **Tables** for detailed numerical data

### Best Practices:
1. **Be specific** about what you want to visualize
2. **Use time words** (daily, weekly, trends) for time-series
3. **Request comparisons** for bar/column charts
4. **Ask for percentages** to get pie charts
5. **Use "relationship" or "correlation"** for scatter plots
6. **Combine with AI** using `--to-ai` for insights

## 🎯 Key Takeaways

- Calliope automatically generates appropriate visualizations
- Chart type is determined by query structure and data
- Use descriptive language to guide visualization choice
- Combine visualizations with AI analysis for deeper insights
