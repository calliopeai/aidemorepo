### Pro Tips:

1. **Be Specific**: The more detail in your request, the better the result
2. **Iterate**: Start simple, then ask for refinements
3. **Mix Models**: Different models excel at different visualization types
4. **Request Interactivity**: Ask for "interactive" or "with plotly" for web-ready charts
5. **Specify Aesthetics**: Mention color schemes, themes, and styling preferences
