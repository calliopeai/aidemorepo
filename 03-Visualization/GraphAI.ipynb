```python
import numpy as np
import matplotlib.pyplot as plt

# Random seed for reproducibility
rng = np.random.default_rng(42)

# ----------------------------
# Assumptions (playful model)
# ----------------------------
# Typical adult woodchuck mass (kg): ~2–6
mass_mean = 4.0
mass_sd = 0.6
mass_min, mass_max = 2.0, 6.0

# Base daily chucking rate per kg of body mass (kg of wood per kg body per day)
base_rate_per_kg = 0.7

# Motivation factor ~ Beta scaled to [0.6, 1.4]
mot_alpha, mot_beta = 2.3, 2.0
def sample_motivation(size):
    return 0.6 + 0.8 * rng.beta(mot_alpha, mot_beta, size=size)

motivation_mean = 0.6 + 0.8 * (mot_alpha / (mot_alpha + mot_beta))

# Wood density (kg/m^3): softwoods ~ 400–550, hardwoods ~ 600–900
density_min, density_max = 450.0, 850.0

# Hardness/softness factor: output scales with sqrt(650 / density)
def softness_factor(density):
    return np.sqrt(650.0 / density)

# Multiplicative noise (lognormal) with mean 1.0
ln_sigma = 0.25
ln_mu = -0.5 * (ln_sigma ** 2)  # ensures E[lognormal]=1
def sample_noise(size):
    return rng.lognormal(mean=ln_mu, sigma=ln_sigma, size=size)

# "Standard log" geometry for fun conversion (0.5 m length, 8 cm diameter)
log_length_m = 0.5
log_radius_m = 0.04
log_volume_m3 = np.pi * (log_radius_m ** 2) * log_length_m
# Assume reference density 650 kg/m^3 for "standard log"
ref_density = 650.0
mass_per_log = ref_density * log_volume_m3  # kg per log

# Unit conversions
kg_to_lb = 2.2046226218

# ----------------------------
# Monte Carlo: "How much wood...?"
# ----------------------------
N = 20000
masses = np.clip(rng.normal(mass_mean, mass_sd, size=N), mass_min, mass_max)
motivations = sample_motivation(N)
densities = rng.uniform(density_min, density_max, size=N)
softness = softness_factor(densities)
noise = sample_noise(N)

daily_kg = base_rate_per_kg * masses * motivations * softness * noise
daily_kg = np.clip(daily_kg, 0, None)
daily_lb = daily_kg * kg_to_lb

# Stats
mean_kg = float(np.mean(daily_kg))
median_kg = float(np.median(daily_kg))
p05_kg, p95_kg = [float(x) for x in np.percentile(daily_kg, [5, 95])]
mean_lb = mean_kg * kg_to_lb
median_lb = median_kg * kg_to_lb

# Helper for secondary axis: kg <-> logs
def kg_to_logs(x):
    return x / mass_per_log

def logs_to_kg(x):
    return x * mass_per_log

# ----------------------------
# Time series over a "season"
# ----------------------------
days = 120
# Pick one representative woodchuck
rep_mass = float(np.clip(rng.normal(mass_mean, mass_sd), mass_min, mass_max))
rep_densities = rng.uniform(density_min, density_max, size=days)
rep_motivations = sample_motivation(days)
rep_softness = softness_factor(rep_densities)
rep_noise = sample_noise(days)
rep_daily_kg = base_rate_per_kg * rep_mass * rep_motivations * rep_softness * rep_noise
rep_daily_kg = np.clip(rep_daily_kg, 0, None)
rep_cum_kg = np.cumsum(rep_daily_kg)
rep_daily_lb = rep_daily_kg * kg_to_lb
rep_cum_lb = rep_cum_kg * kg_to_lb

# ----------------------------
# Sensitivity: mass vs density (expected daily kg)
# ----------------------------
mass_grid = np.linspace(2.0, 6.0, 60)
density_grid = np.linspace(450.0, 850.0, 80)
M, D = np.meshgrid(mass_grid, density_grid, indexing='xy')
expected_daily_kg = base_rate_per_kg * M * softness_factor(D) * motivation_mean  # noise mean = 1

# ----------------------------
# Plotting
# ----------------------------
plt.rcParams.update({"figure.figsize": (13, 9), "axes.grid": True})

fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(2, 2)

# 1) Distribution histogram
ax1 = fig.add_subplot(gs[0, 0])
bins = 50
ax1.hist(daily_kg, bins=bins, color="#5DA5DA", alpha=0.85, edgecolor="white")
ax1.set_title("How much wood can a woodchuck chuck? (Simulated daily output)")
ax1.set_xlabel("Daily wood chucked (kg)")
ax1.set_ylabel("Frequency")

# Secondary x-axis: "logs"
secax = ax1.secondary_xaxis('top', functions=(kg_to_logs, logs_to_kg))
secax.set_xlabel("Approximate logs per day (0.5 m × 8 cm, ref 650 kg/m³)")

# Vertical lines for stats
ax1.axvline(mean_kg, color="#F17CB0", linewidth=2, linestyle="--", label=f"Mean: {mean_kg:.2f} kg ({mean_lb:.1f} lb)")
ax1.axvline(median_kg, color="#B2912F", linewidth=2, linestyle=":", label=f"Median: {median_kg:.2f} kg ({median_lb:.1f} lb)")
ax1.fill_betweenx([0, ax1.get_ylim()[1]], p05_kg, p95_kg, color="#60BD68", alpha=0.12, label=f"90% range: {p05_kg:.2f}–{p95_kg:.2f} kg")

ax1.legend(loc="upper right", frameon=True)

# Fun annotation
ax1.text(
    0.02, 0.97,
    "Answer: as much wood as a woodchuck could chuck\nif a woodchuck could chuck wood.",
    transform=ax1.transAxes,
    va="top", ha="left", fontsize=9, color="#444444"
)

# 2) Season time series
ax2 = fig.add_subplot(gs[0, 1])
t = np.arange(1, days + 1)

color_daily = "#4E79A7"
color_cum = "#E15759"

ax2.plot(t, rep_daily_kg, color=color_daily, label="Daily (kg)")
ax2.set_title("One woodchuck over a season (simulated)")
ax2.set_xlabel("Day")
ax2.set_ylabel("Daily wood (kg)", color=color_daily)
ax2.tick_params(axis='y', labelcolor=color_daily)

ax2b = ax2.twinx()
ax2b.plot(t, rep_cum_kg, color=color_cum, linewidth=2.0, label="Cumulative (kg)")
ax2b.set_ylabel("Cumulative wood (kg)", color=color_cum)
ax2b.tick_params(axis='y', labelcolor=color_cum)

# Small legend
lines1, labels1 = ax2.get_legend_handles_labels()
lines2, labels2 = ax2b.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1 + labels2, loc="upper left", frameon=True)

# 3) Sensitivity heatmap (mass vs density)
ax3 = fig.add_subplot(gs[1, :])
cmap = plt.cm.viridis
im = ax3.pcolormesh(mass_grid, density_grid, expected_daily_kg.T, shading='auto', cmap=cmap)
cb = plt.colorbar(im, ax=ax3)
cb.set_label("Expected daily wood (kg)")

ax3.set_title("Sensitivity: body mass and wood density vs expected daily output")
ax3.set_xlabel("Woodchuck mass (kg)")
ax3.set_ylabel("Wood density (kg/m³)")

# Overlay example contours
cs = ax3.contour(mass_grid, density_grid, expected_daily_kg.T, colors='white', linewidths=0.7, levels=6)
ax3.clabel(cs, inline=True, fontsize=8, fmt="%.2f kg")

# Footer note
fig.text(
    0.5, 0.01,
    "This is a playful simulation using arbitrary-but-plausible assumptions; not a biological measurement.",
    ha="center", va="bottom", fontsize=9, color="#555555"
)

plt.show()
```
