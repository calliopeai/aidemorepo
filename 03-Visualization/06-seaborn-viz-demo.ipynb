```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Sample data
data = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [23, 45, 56, 78]
})

# Line Plot
plt.figure(figsize=(8, 4))
plt.plot(data['Category'], data['Values'], marker='o')
plt.title('Line Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.grid(True)
plt.show()

# Bar Plot
plt.figure(figsize=(8, 4))
sns.barplot(x='Category', y='Values', data=data, palette='viridis')
plt.title('Bar Plot')
plt.show()

# Histogram
plt.figure(figsize=(8, 4))
data = np.random.randn(1000)
plt.hist(data, bins=30, alpha=0.7, color='green')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Scatter Plot
plt.figure(figsize=(8, 4))
x = np.random.rand(100)
y = np.random.rand(100)
plt.scatter(x, y, c='blue', alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# Heatmap
data = np.random.rand(10, 10)
plt.figure(figsize=(8, 4))
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()

# Box Plot
data = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C'], size=100),
    'Values': np.random.randn(100)
})
plt.figure(figsize=(8, 4))
sns.boxplot(x='Category', y='Values', data=data)
plt.title('Box Plot')
plt.show()

# Pair Plot
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px
import plotly.graph_objects as go
from wordcloud import WordCloud
import networkx as nx

# Load dataset
df = sns.load_dataset('iris')

# Set the style for seaborn
sns.set(style="whitegrid")

# 1. Pairplot using seaborn
sns.pairplot(df, hue='species', height=2.5)
plt.suptitle('Pairplot of Iris Dataset', y=1.02)

# 2. 3D Scatter Plot using matplotlib
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
colors = {'setosa': 'r', 'versicolor': 'g', 'virginica': 'b'}
ax.scatter(df['sepal_length'], df['sepal_width'], df['petal_length'], c=df['species'].map(colors))
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')
plt.title('3D Scatter Plot of Iris Dataset')
plt.show()

# 3. Heatmap using seaborn
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Iris Dataset')
plt.show()

# 4. Interactive Scatter Plot using Plotly
fig = px.scatter(df, x='sepal_length', y='petal_length', color='species', 
                 size='sepal_width', hover_data=['petal_width'], title='Interactive Scatter Plot')
fig.show()

# 5. Sunburst Chart using Plotly
fig = px.sunburst(df, path=['species', 'sepal_length'], values='petal_length', title='Sunburst Chart of Iris Dataset')
fig.show()

# 6. Violin Plot using seaborn
plt.figure(figsize=(8, 6))
sns.violinplot(x='species', y='sepal_length', data=df, inner='quartile')
plt.title('Violin Plot of Sepal Length by Species')
plt.show()

# 7. Word Cloud using WordCloud
text = " ".join(species for species in df.species)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Iris Species')
plt.show()

# 8. Network Graph using NetworkX
G = nx.cycle_graph(5)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='skyblue', edge_color='gray', node_size=2000, alpha=0.7)
plt.title('Simple Network Graph')
plt.show()

# 9. Density Plot using seaborn
plt.figure(figsize=(8, 6))
sns.kdeplot(df['sepal_length'], shade=True, color="b", label='Sepal Length')
sns.kdeplot(df['petal_length'], shade=True, color="r", label='Petal Length')
plt.title('Density Plot of Sepal and Petal Length')
plt.legend()
plt.show()

# 10. Parallel Coordinates using Pandas Plotting
from pandas.plotting import parallel_coordinates
plt.figure(figsize=(10, 6))
parallel_coordinates(df, 'species', colormap='Dark2')
plt.title('Parallel Coordinates Plot of Iris Dataset')
plt.show()
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

# Sample data
df = pd.DataFrame({
    "Fruit": ["Apples", "Oranges", "Bananas", "Apples", "Oranges", "Bananas"],
    "Amount": [4, 1, 2, 2, 4, 5],
    "City": ["SF", "SF", "SF", "NYC", "NYC", "NYC"]
})

# Create a Plotly figure
fig = px.bar(df, x="Fruit", y="Amount", color="City", barmode="group")

# Initialize Dash app
app = dash.Dash(__name__)

# Define app layout
app.layout = html.Div(children=[
    html.H1(children='Hello Dash'),
    dcc.Graph(id='example-graph', figure=fig)
])

# Run server
if __name__ == '__main__':
    app.run_server(debug=True, port=8055)
```
