## 📚 Command Quick Reference

### Essential Commands

```python
# Get help
%%calliope help

# List models
%%calliope list-models

# Natural language query (generates SQL, executes, visualizes)
%%calliope ask-sql datasource_id
What are the top 10 customers by revenue?

# Generate SQL only (no execution)
%%calliope generate-sql datasource_id
Show monthly sales trends

# Run custom SQL directly
%%calliope run-sql datasource_id
SELECT * FROM customers LIMIT 10

# Train with business logic
%%calliope rag-train datasource_id
{
    "documentation": ["ARR = Annual Recurring Revenue"],
    "sql": ["SELECT SUM(revenue*12) FROM subscriptions"]
}

# Get follow-up question suggestions
%%calliope followup-questions datasource_id
```
