## 💡 Key Calliope Syntax

### Basic:
```python
%%calliope ask model_name --format code
Your natural language request here
```

### With Column Cleaning (Recommended - Used in All Notebooks):
```python
%%calliope ask gpto --format code
Load ~/lab/data/CSVs/file.csv

Clean column names:
1. Strip whitespace: df.columns = df.columns.str.strip()
2. Replace spaces: df.columns = df.columns.str.replace(' ', '_')
3. Lowercase: df.columns = df.columns.str.lower()

Then do analysis...
```

### Output Redirection:
```python
%%calliope ask gpto --format code
Calculate stats and store as:
- result_df: the processed dataframe
- top_10: list of top 10 items
- summary_dict: dictionary of key metrics

Print all three.
```

## 📝 Tips for Success

1. **Always clean column names first** - Helps AI generate better code
2. **Be explicit about output** - Tell AI exactly what to store
3. **Reference previous cells** - Build complex analyses step-by-step
4. **Sample large datasets** - 50k-100k rows for performance
5. **Use appropriate models:**
   - Simple tasks → Fast models (gemini-flash, haiku, gpto)
   - Complex analysis → Powerful models (opus, gpt-5, sonnet-4.5)
6. **Install libraries as needed** - Use `sudo pip install` in cells
7. **Export interactive reports** - Save plotly as HTML
