```python
import importlib

libraries = [
    ("pandas", "pandas"),
    ("numpy", "numpy"),
    ("scipy", "scipy"),
    ("statsmodels", "statsmodels"),
    ("matplotlib", "matplotlib"),
    ("seaborn", "seaborn"),
    ("plotly", "plotly"),
    ("bokeh", "bokeh"),
    ("altair", "altair"),
    ("sklearn", "scikit-learn"),
    ("xgboost", "xgboost"),
    ("lightgbm", "lightgbm"),
    ("geopandas", "geopandas"),
    ("folium", "folium"),
    ("ipyleaflet", "ipyleaflet"),
    ("networkx", "networkx"),
    ("igraph", "igraph"),
    ("prophet", "prophet"),
    ("pmdarima", "pmdarima"),
    ("shap", "shap"),
    ("yellowbrick", "yellowbrick"),
    ("wordcloud", "wordcloud"),
    ("missingno", "missingno"),
    ("ipywidgets", "ipywidgets"),
    ("bqplot", "bqplot"),
    ("holoviews", "holoviews"),
    ("hvplot", "hvplot"),
    ("datashader", "datashader")
]

for library, module_name in libraries:
    try:
        module = importlib.import_module(module_name)
        version = getattr(module, '__version__', 'Unknown')
        print(f'{library}: {version}')
    except ImportError:
        print(f'{library}: Not Installed')
```
