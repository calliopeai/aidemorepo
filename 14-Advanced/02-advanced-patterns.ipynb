```python
%%calliope ask openai-chat:gpt-4o --format code
Pipeline Step 1: Load and Clean

Load ~/lab/data/CSVs/MVA_Vehicle_Sales_Counts_by_Month_for_Calendar_Year_2002_through_February_2025.csv
Clean the data and store as 'vehicle_df'
Calculate and store:
- total_years: number of unique years
- date_range: tuple of (min_date, max_date)

Print summary statistics and the stored variables
```

```python
import pandas as pd

# Load the data
vehicle_df = pd.read_csv('~/~/lab/data/CSVs/MVA_Vehicle_Sales_Counts_by_Month_for_Calendar_Year_2002_through_February_2025.csv')

# Clean the data
vehicle_df.dropna(inplace=True)

# Convert date column to datetime if necessary, assume the column is named 'Date'
vehicle_df['Date'] = pd.to_datetime(vehicle_df['Date'])

# Calculate total years
total_years = vehicle_df['Date'].dt.year.nunique()

# Calculate date range
min_date = vehicle_df['Date'].min()
max_date = vehicle_df['Date'].max()
date_range = (min_date, max_date)

# Print summary statistics and the stored variables
print(vehicle_df.describe())
print(f"Total years: {total_years}")
print(f"Date range: {date_range}")
```

## Summary: Best Practices

### Output Redirection:
1. **Be explicit**: Tell AI exactly what to store: "store as 'variable_name'"
2. **Use descriptive names**: `obesity_by_state` not just `data1`
3. **Print for verification**: Always print stored variables to confirm

### Variable Interpolation:
1. **Reference by name**: "Using variable_name from previous cell"
2. **List what you're using**: Make it clear which variables you need
3. **Chain logically**: Build complex analyses step-by-step

### Model Selection:
- **Quick analysis**: gemini-flash, claude-haiku
- **Standard work**: gpto, gpt-4o, claude-sonnet
- **Complex logic**: gpt-5, claude-opus
- **Specific tasks**: cohere (structured), mistral (Europe)

### Performance:
1. Sample large datasets (50K-100K rows)
2. Use fast models for simple tasks
3. Chain operations rather than one mega-request
4. Store intermediate results for reuse
